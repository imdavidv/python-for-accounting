```python
# Import libraries
import pandas as pd
```

```python
# Import the data files for review
customers=pd.read_excel('Sales_Customers.xlsx', sheet_name='Sheet1')
orders=pd.read_excel('Sales_Orders.xlsx', sheet_name='Sheet1')
```

```python
# Quick Preview of data frame
customers.head()
```

```python
# Quick Preview of data frame
orders.head()
```

```python
# Create a new data frame and merge the previous two
results=pd.merge(customers,orders, on='custid', how='outer', left_index=False, right_index=False, indicator='indicator_column')
```

```python
# Quick Preview of data frame
results.head(10)
```
