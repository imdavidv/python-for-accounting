# Daily Bank Transaction Review

```python
#First we will import the pandas library
import pandas as pd
```

```python
#Then we will select the two files to compare

checks = pd.read_excel('BankRec_example.xlsx')
bank = pd.read_excel('BankRec_BankofAmerica.xlsx')
```

```python
# Quick review of the 'checks' data frame
checks.head()
```

```python
# Quick review of the 'bank' data frame
bank.head()
```

```python
# Now we will create a third data frame and merge the previous data frame's into one.
# Similar to SQL, I can join the two data frames by using a 'left, right, outer, and inner'
# I can also add an indicator that will let me know which data frame the data EXISTS or NOT EXISTS

review = pd.merge(checks, bank, how='outer', left_on='CheckNo', right_on='CheckCleared', indicator='location')
review
```

```python
# 'unknown' will be used to indicate transactions that do not exist in the check register
unknown = review.loc[review['location'] == 'right_only']

# A quick preview
unknown.head()
```

```python
# 'known' will be used to indicate transactions that do exist in the check register
known = review.loc[review['location'] == 'both']

# I will also add a column that will indicate if any difference exists between the actual check and the transaction
# amount from the bank
known['Difference'] = known['TotalDue'] - known['AmountCleared']

# A quick preview
known.head(15)
```

```python
known.info()
```

```python
# Now I will create another data frame to show only checks that cleared with a different amount
DiffAmount = known.loc[known['Difference'] != 0]
DiffAmount.head()
```
